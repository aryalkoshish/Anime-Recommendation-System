```python
import spacy
import re
from collections import Counter
from wordcloud import WordCloud
from textblob import TextBlob
# Remove stopwords, punctuation, and symbols
nlp = spacy.load("en_core_web_sm")

def preprocess(sentence):
    if pd.isnull(sentence):
        return ""
    doc = nlp(sentence)
    filtered_words = []
    for token in doc:
        if not token.is_stop and not token.is_punct and not re.match('^[^a-zA-Z0-9]+$', token.text):
            filtered_words.append(token.lemma_)
    
    # Join the words back into a sentence
    return " ".join(filtered_words)
```

```python
# Apply preprocessing to the genre column
final_df['clean_genre'] = final_df['genre'].apply(preprocess)
```

```python
# Displaying all the cleaned text in the genre column
text = " ".join(line for line in final_df['clean_genre'])
text
```

```python
# Creating a word cloud for the 'clean_genre' column
wordcloud_text = WordCloud(background_color = 'white', max_words = 100).generate(' '.join(final_df['clean_genre']))

# Displaying the word cloud
plt.figure(figsize = (10, 10))
plt.imshow(wordcloud_text, interpolation = 'bilinear')
plt.axis('off')
plt.title('Word Cloud for Genre')
plt.show()
```

```python
# Display the top 10 genres based on count 
topwords = Counter(" ".join(final_df['clean_genre']).split()).most_common(10)
topwords
```

```python
# Show the top 5 genres depending on the user's gender  
genders = ['Male', 'Female', 'Non-Binary']

top_genres_by_gender = {}

for gender in genders:
    subset = final_df[final_df['gender'] == gender]
    count = Counter(" ".join(subset['clean_genre']).split()).most_common(5)
    top_genres_by_gender[gender] = count

# Print the top genres for each gender
for gender, top_genres in top_genres_by_gender.items():
    print(f"Top genres for \033[1m{gender}\033[0m:")
    for genre, count in top_genres:
        print(f"Genre: \033[1m{genre}\033[0m, Count: \033[1m{count}\033[0m")
    print()
```
